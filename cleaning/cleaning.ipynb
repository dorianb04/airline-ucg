```python
companies = ['air-france', 'asiana-airlines', 'british-airways', 'iberia', 'japan-airlines-jal', 'lufthansa',
             'or520-emirates', 'or1030-korean-air', 'qatar-airways', 'ryanair' ,'singapore_airlines', 
             'swiss-international-air-lines-swiss','turkish-airlines']

numeric_columns = ['legroom_rating', 'seat_comfort_rating', 'entertainment_rating',
                   'customer_service_rating', 'value_rating', 'cleanliness_rating',
                   'check_in_rating', 'food_rating']
```

```python
def clean_data(csv, i):
    df = pd.read_csv(csv)
    df = df.dropna(subset=['review_text', 'rating'])     # delete lines where review text or rating is empty 
    df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean()) # replace the missing values with the mean of the column
    df['travel_date'] = pd.to_datetime(df['travel_date'], format='%B %Y', errors='coerce') # convert the date columns to datetime
    df['travel_date'] = pd.to_datetime(df['travel_date'], errors='coerce')
    df['travel_date'] = df['travel_date'].dt.to_period('M')  # Conserve uniquement mois et annÃ©e
    df['review_date'] = pd.to_datetime(df['review_date'], format='%B %d, %Y', errors='coerce')
    df = df.dropna(subset=['travel_date', 'review_date']) # delete lines where travel date or review date is empty
    df = df.drop_duplicates(subset=['review_text', 'review_date'], keep='first') # delete duplicates
    df = df[(df['rating'] >= 0) & (df['rating'] <= 5)]  # delete lines where rating is not between 0 and 5
    df.to_csv('airline-ucg/cleaning/cleaned_data_csv/'+companies[i]+'_reviews_cleaned.csv', index=False)
    return df
```

```python
for i in range(len(companies)):
    csv = 'airline-ucg/crawling/data_csv/'+companies[i]+'_reviews.csv'
    clean_data(csv, i)
```
